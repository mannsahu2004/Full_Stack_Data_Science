Swap 2-variable in python


```python
a = 5
b = 6
```

```python
a = b
b = a
```

```python
print(a)
print(b)
```

```python
a1 = 7
b1 = 8
```

```python
temp = a1
a1 = b1
b1 = temp
```

```python
print(a1)
print(b1)
```

```python
# swap variable without using third variable
a2 = 5
b2 = 6
```

```python
a2 = a2+b2
b2 = a2-b2
a2 = a2-b2
```

```python
print(a2)
print(b2)
```

```python
a2,b2 = b2,a2
```

```python
print(a2)
print(b2)
```
