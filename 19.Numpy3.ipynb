```python
#Numpy(array) -- fast ,best for numerical data
import numpy as np
marks1 = np.array([80,75,90,60])
```

Total

```python
#Python -- more code
total = 0 
for m in marks :
    total += m
print(total)
```

```python
#Numpy -- work in one line 
np.sum(marks1)
```

Average(mean)

```python
avg = sum(marks)/ len(marks)
print(avg)
```

```python
np.mean(marks1) # In numpy there is already builtin functions
```

```python
new_marks = []
for m in marks:
    new_marks.append(m+5)
print(new_marks)
```

Sum of Rows and Columns

```python
#Complex
data = [[10,20,30],
        [40,50,60]]

#Initial values -- three zeroes for three columns
col_sum = [0,0,0]
print("Start col_sum:", col_sum)

for row in data:
    print("Current row:",row)
    
    for i in range(3):
        col_sum[i] += row[i]
        print(f"After adding row[{i}] = {row[i]} → col_sum:", col_sum)

print("Final column-wise sum:",col_sum)
```

```python
#Axis concept --- fast & readable
data1 = np.array([[10,20,30], 
                 [40,50,60]])
print(np.sum(data1, axis=0)) #column-wise
print(np.sum(data1, axis=1)) #row-wise
```
